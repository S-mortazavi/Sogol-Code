```python
import finpy_tse as tse
df=tse.Get_Price_History(stock='وبانک',start_date='1401-01-01',
                          end_date='1402-08-02',ignore_date=False,
                          adjust_price=True,show_weekday=True,double_date=True)
display(df)
```

```python
import pytse_client as pytse
ticker=pytse.Ticker('وبانک')
```

```python
print('درصد سهام شناور: \n',ticker.float_shares)
print('تعداد سهام: \n',ticker.total_shares)
print('حجم مبنا: \n',ticker.base_volume)
print('نسبت p/e: \n',ticker.p_e_ratio)
```

```python
print(' درصد سهامداران عمده:')
display(ticker.shareholders) 
```

```python
import numpy as np
import talib
adj_close=df.iloc[:,16]
print('Adj Close: \n',adj_close,'\n\n')
EMA=talib.EMA(adj_close,timeperiod=10)
print('EMA: \n',EMA,'\n\n')
EMA.values
```

```python
MACD=talib.MACD(adj_close)
print('MACD: \n',MACD)
```

```python
adj_high=df.iloc[:,14]
print('Adj High \n',adj_high,'\n\n')
adj_low=df.iloc[:,15]
print('Adj Low \n',adj_low,'\n\n')

PSAR=talib.SAR(high=adj_high,low=adj_low)
print('PSAR: \n',PSAR)
```

```python
volume=df.iloc[:,7]
print('volume: \n',volume,'\n\n')
OBV=talib.OBV(adj_close,volume)
print('OBV: \n',OBV,'\n\n')
```
