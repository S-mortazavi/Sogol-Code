```python
import numpy as np

#first answer:

def fun1(array,number):
    count=np.sum(array==number)
    if count>0:
        print(f'this number appears {count} times in array!')
    else:
        print('-1')

print('Example 1:')
array=np.array([[2,3,2,9],[7,2,4,3]])
number=2
fun1(array,number)

print('Example 2:')
array=np.array([[4,7,9,12],[8,4,4,13]])
number=2
fun1(array,number)

#second answer:

def fun2(array,number):
    indices=np.where(array==number)
    count=indices[0].size
    if count>0:
        print(f'this number appears {count} times in array!')
    else:
        print('-1')

print('Example 3:')
array=np.array([[4,3,9,11,2],[9,8,3,4,14]])
number=1
fun2(array,number)
print('Example 4:')
array=np.array([[2,3],[7,2]])
number=2
fun2(array,number)
```

```python
import numpy as np
a=np.array([1,1,1])
b=np.array([3,2,0])
c=np.array([4,2,9])
d=np.vstack((a,b,c)).T
print(d)

inv_d=np.linalg.inv(d)
print('inverse matrix: \n',inv_d)
```

```python
stock_price=np.array([100,105,104,103,106,105,107,109,108,110])
signal=np.where(stock_price[1:]>stock_price[:-1],1,0)
signal=np.append(signal,np.nan)
price=np.vstack((stock_price,signal))
print(price)
```

```python
stock_price=np.array([100,105,104,103,106,105,107,109,108,110])
price_2=np.roll(stock_price,-2)
signal=np.where(stock_price<price_2,1,0).astype(float)
signal[-2:]=np.nan
price_and_signal=np.vstack((stock_price,signal))
print(price_and_signal)
```

```python
import pandas as pd
import finpy_tse as tse
df=tse.Get_Price_History(stock='',start_date='1401-01-01',end_date='1402-09-10',
                            ignore_date=False,adjust_price=True,show_weekday=True,double_date=True)
display(df)
df1=df.drop(["Weekday","Open","High","Low","Close","Final","Volume","Value","No","Ticker","Name","Market","Adj Open",
        "Adj High","Adj Low","Adj Final"], axis=1)
display(df1)
```

```python
print(df1.shape[0])
print('max: ',df1["Adj Close"].max())
print('min: ',df1["Adj Close"].min())
print('mean: ',df1["Adj Close"].mean())
print('standard deviation:',df1["Adj Close"].std())
df1.describe()
```

```python
import talib
adj_close=df1.iloc[:,1]
EMA=talib.EMA(adj_close,timeperiod=10)
EMA=pd.DataFrame(EMA)
display(EMA)
df1["EMA"]=EMA
df1=df1.dropna(axis=0)
EMA=EMA.dropna(axis=0)
display(df1)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
adj_close=df1.iloc[:,1]
time=df1.iloc[:,0]

fig, ax=plt.subplots(figsize=(12, 5))
ax.plot(time,adj_close,color='blue')
ax.plot(time,EMA,color='red')
ax.set_title("Price and Exponential Moving Average chart",fontname='Times New Roman', fontsize=15)
ax.set_xlabel("Time",fontname='Times New Roman', fontsize=15)
ax.set_ylabel("Price & EMA",fontname='Times New Roman', fontsize=15)

ax.yaxis.set_major_locator (ticker.MultipleLocator(10000))
ax.yaxis.set_minor_locator (ticker.MultipleLocator(1000))

ax.legend(['Price', 'EMA'])
```

```python
df1['Buy Signal']=np.where((df1['Adj Close']>df1['EMA'])&(df1['Adj Close'].shift(1)<=df1['EMA'].shift(1)),'Buy','')
df1['Sell Signal']=np.where((df1['Adj Close']<df1['EMA'])&(df1['Adj Close'].shift(1)>=df1['EMA'].shift(1)),'Sell','')
df1['Signal']=np.where(df1['Buy Signal']=='Buy','Buy',np.where(df1['Sell Signal']=='Sell','Sell',''))
df_signal=df1[['Date','Signal']]
display(df_signal.head(50))
```

```python
df1 = df1.fillna('')
df2 = df1[['Signal']] 
df2 = df2[df2['Signal'] != '']
display(df2)
```

```python
df3 =df1[["Signal","Adj Close"]]
df3 = df3[df3['Signal'] != '']
df3 = df3[df3['Adj Close'] != '']
df3['return'] = df3[['Adj Close']].pct_change()
df3['asset'] = np.nan
df3.loc['1401-02-04', 'asset'] = 100
for i in range(2, len(df3)):
    if df3['Signal'].iloc[i] == 'Buy':
        df3['asset'].iloc[i] = df3['asset'].iloc[i-1]
    if df3['Signal'].iloc[i] == 'Sell':
        df3['asset'].iloc[i] = df3['asset'].iloc[i-1] * (1 + df3['return'].iloc[i])
        
df3['asset'] = df3['asset']
display(df3)
```

```python
df4 = df3[["Adj Close", "Signal", "asset"]]
df4['Transaction'] = np.where(df4['Signal']=='Buy', df4['asset']*(0.4/100), df4['asset']*(0.8/100)).round(2)
display(df4)
```

```python
stock_return = ((df4['asset'].iloc[-1]-df4['Transaction'].sum())-100)/100
print(stock_return*100,'%')
```
